```python
import json
import os
import urllib
import requests
import re

path = './vn_register_webportal/data/'
os.listdir(path)
```

```python
def download_file(url, save_dir, save_name):
    response = requests.get(url)
    save_path = os.path.join(save_dir, save_name)
    with open(save_path, 'wb') as file:
        file.write(response.content)
```

```python
count = {
    "doc": 0,
    "docx": 0, 
    "zip": 0,
    "rar": 0,
    "pdf": 0
}
```

```python
not_found_url_page = []
for fname in os.listdir(path):
    data = json.load(open(os.path.join(path, fname), 'r'))
    save_dir = f"./attachments/{fname.rstrip('.json')}"
    if not os.path.exists(save_dir):
        os.mkdir(save_dir)
    for item in data:
        try:
            if item['attachment_url'] is not None:
                url = item['attachment_url']
                if not url.startswith('http'): url = 'http://' + url
                # print('URL:', url)
                filename = re.search(r'[^\/]+$', url).group(0)
                file_extension = re.search(r'\.([a-zA-Z0-9]+)$', filename).group(1).lower()
                count[file_extension] += 1
                # if not os.path.exists(os.path.join(save_dir, file_extension)):
                #     os.mkdir(os.path.join(save_dir, file_extension))
                # download_file(url=url, save_dir=os.path.join(save_dir, file_extension), save_name=f"{item['item_id']}-{filename}")
                
        except Exception as e:
            print(e)
            not_found_url_page.append((item['item_id'], item['page_url']))
```
